```python
import torch
from torchcubicspline import(natural_cubic_spline_coeffs, 
                             NaturalCubicSpline)

length, channels = 7, 2
t = torch.linspace(0, 1, length)
x = torch.rand(length, channels)
coeffs = natural_cubic_spline_coeffs(t, x)
spline = NaturalCubicSpline(coeffs)
points = torch.linspace(0, 1, 100)
out = spline.evaluate(points)


# plotting
import matplotlib.pyplot as plt
plt.plot( x[:,0].numpy(),x[:,1].numpy(), 'o', label='data')
plt.plot(out[:,0].numpy(), out[:,1].numpy(), label='spline')
plt.legend()
plt.show()
```
